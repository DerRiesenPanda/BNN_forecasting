```python
def split_sequences(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
        end_ix = i + n_steps
        if end_ix >= len(sequences):
            break
        seq_x, seq_y = sequences[i:end_ix], sequences[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)


def multi_time(data, n_input, n_out):

    #train = data.reshape((data.shape[0]*data.shape[1], data.shape[2]))
    X,y = list(), list()
    in_start= 0

    for i in range(len(data)):

        in_end = in_start + n_input
        out_end = in_end + n_out
        if out_end <= len(data):
            X.append(data[in_start:in_end])
            y.append(data[in_end:out_end])
            in_start +=1
    return np.array(X), np.array(y)
```

```python

class BNN_onestep(PyroModule):
    def __init__(self, in_dim=1, out_dim=1,prior_scale=0):
        super().__init__()
        assert in_dim > 0 and out_dim > 0
        #self.layer_sizes = [in_dim] + n_hid_layers * [hid_dim] + [out_dim]
        self.layer_1 = PyroModule[nn.Conv1d](in_channels = 3, out_channels = 6, kernel_size = 1, padding = 'same')
        self.activation = PyroModule[nn.ReLU]()
        self.layer_2 = PyroModule[nn.Conv1d](6, 12, 1, padding='same')
        self.linear_1 = PyroModule[nn.Linear](12,1)

        self.layer_1.weight = PyroSample(dist.Normal(0, 1).expand([6,3,1]).to_event(3))
        self.layer_1.bias = PyroSample(dist.Normal(0, 1).expand([6]).to_event(1))
        self.layer_2.weight = PyroSample(dist.Normal(0, 1).expand([12,6,1]).to_event(3))
        self.layer_2.bias = PyroSample(dist.Normal(0, 1).expand([12]).to_event(1))
        self.linear_1.weight = PyroSample(dist.Normal(0, 1).expand([1,12]).to_event(2))
        self.linear_1.bias = PyroSample(dist.Normal(0, 1).expand([1]).to_event(1))

    def forward(self, x, y = None):

        x = self.activation(self.layer_1(x))
        x = self.activation(self.layer_2(x))
        x = torch.flatten(x,1)
        x = self.activation(self.linear_1(x))
        mu = x
        #print(mu.shape)
        sigma = pyro.sample("sigma", dist.Gamma(.5, 1))

        if y is not None:
            y = torch.flatten(y,1)

        with pyro.plate("data", x.shape[0]):

            obs = pyro.sample("obs", dist.Normal(mu, sigma * sigma).to_event(1), obs = y)
            #print(obs)
            #print(obs.shape)
        return mu
```

```python

pyro.clear_param_store()
model = BNN_onestep()
mean_field_guide = AutoDiagonalNormal(model)
optimizer = pyro.optim.Adam({"lr": 0.01})

svi = SVI(model, mean_field_guide, optimizer, loss=Trace_ELBO())
pyro.clear_param_store()

num_epochs = 1000
progress_bar = trange(num_epochs)


for epoch in progress_bar:
    loss = 0
    for batch_id, data in enumerate(train_loader):
        loss += svi.step(data[0], data[1])
    progress_bar.set_postfix(loss=f"{loss / len(x_train):.3f}")
```

```python

predictive = Predictive(model, guide=mean_field_guide, num_samples=500)
#preds = predictive(x_test)
pred_cnn= []
plus_error_cnn = []
minus_error_cnn = []
real = []
inputs = []

for batch_id, data in enumerate(test_loader):
        preds = predictive(data[0])
        y_std = preds['obs'][:,0,:].detach().numpy().std(axis=0, ddof=1)
        y_preds = preds['obs'][:,0,:].detach().numpy().sum(axis=0)/500
        y_real = data[1]
        real.append(y_real)
        inputs.append(data[0])
        plus_error_cnn.append(y_preds - 2* y_std)
        minus_error_cnn.append(y_preds + 2* y_std)
        pred_cnn.append(y_preds)
```

```python

class BNN_multistep(PyroModule):
    def __init__(self, in_dim=1, out_dim=1,prior_scale=0):
        super().__init__()
        assert in_dim > 0 and out_dim > 0
        #self.layer_sizes = [in_dim] + n_hid_layers * [hid_dim] + [out_dim]
        self.layer_1 = PyroModule[nn.Conv1d](in_channels = 3, out_channels = 12, kernel_size = 2, padding = 'same')
        self.layer_2 = PyroModule[nn.Conv1d](12, 32, 3, padding='same')
        self.pool = nn.MaxPool1d(2,padding=0)
        self.layer_3 = PyroModule[nn.Conv1d](32, 16, 3, padding='same')

        self.activation = PyroModule[nn.ReLU]()

        self.linear_1 = PyroModule[nn.Linear](16,50)
        self.linear_2 = PyroModule[nn.Linear](50,3)

        self.layer_1.weight = PyroSample(dist.Normal(0, 1).expand([12,3,1]).to_event(3))
        self.layer_1.bias = PyroSample(dist.Normal(0, 1).expand([12]).to_event(1))

        self.layer_2.weight = PyroSample(dist.Normal(0, 1).expand([32,12,1]).to_event(3))
        self.layer_2.bias = PyroSample(dist.Normal(0, 1).expand([32]).to_event(1))

        self.layer_3.weight = PyroSample(dist.Normal(0, 1).expand([16,32,1]).to_event(3))
        self.layer_3.bias = PyroSample(dist.Normal(0, 1).expand([16]).to_event(1))

        self.linear_1.weight = PyroSample(dist.Normal(0, 1).expand([50,16]).to_event(2))
        self.linear_1.bias = PyroSample(dist.Normal(0, 1).expand([50]).to_event(1))

        self.linear_2.weight = PyroSample(dist.Normal(0, 1).expand([3,50]).to_event(2))
        self.linear_2.bias = PyroSample(dist.Normal(0, 1).expand([3]).to_event(1))

    def forward(self, x, y = None):

        x = self.activation(self.layer_1(x))
        x = self.activation(self.layer_2(x))
        x = self.activation(self.layer_3(x))
        x = torch.flatten(x,1)
        x = self.activation(self.linear_1(x))
        x = self.activation(self.linear_2(x))
        mu = x
        #print(mu.shape)
        sigma = pyro.sample("sigma", dist.Gamma(.5, 1))

        if y is not None:
            y = torch.flatten(y,1)

        with pyro.plate("data", x.shape[0]):

            obs = pyro.sample("obs", dist.Normal(mu, sigma * sigma).to_event(1), obs = y)
            #print(obs)
            #print(obs.shape)
        return mu
```
